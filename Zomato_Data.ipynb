```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato data .csv")
```

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)


df['rate'] = df['rate'].apply(handleRate)
print(df.head())
```

```python
sns.countplot(x=df['listed_in(type)'], palette="Set1")
plt.xlabel('Type OF Resturant')
```

```python
Grouped_Data = df.groupby(['listed_in(type)'])['votes'].sum()
Result = pd.DataFrame({'votes': Grouped_Data})
plt.plot(Result, c="red", marker="o")
plt.xlabel('Types of Resturants', c="black", size=10)
plt.ylabel('Number of Votes', c='black', size=10)
```

```python
plt.hist(df['rate'], bins=8)
plt.title('Ratings', size=20)
plt.show()
```

```python
count = df['approx_cost(for two people)']
x = sns.countplot(x=count, palette='Set1')
plt.xlabel('Average amount spend by a couple')
```

Result: Majority of the couple order food of rupees 300-/


```python
def ChangeOrderValue(value):
    if (value == 'Yes'):
        value = "Online"
    elif (value == "No"):
        value = "Offline"
    return value


df['online_order'] = df['online_order'].apply(ChangeOrderValue)

count = df['online_order']
sns.countplot(x=count, palette='Set2')
plt.xlabel("Method OF order food")
```

Result: Majority of the people preffered offline method of buying food

```python
plt.figure(figsize = (6,6))
sns.boxplot(x= 'online_order', y = 'rate', data=df, palette='Set1')
```

Result: People who order food online give better rating as compared to one who is buying food offline

```python
Table = df.pivot_table(index= 'listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(Table, annot=True, cmap='YlGnBu',fmt='d')
plt.title('HEATMAP PLOT', c='red', size=20)
plt.xlabel('Method of Order')
plt.ylabel('Type OF Resturant')
```

Dinnings primarily accept offline orders, whereas cafes receives online orders.
This suggests that clients prefer to place orders in person at resturants, but prefer online ordering at cafes.
